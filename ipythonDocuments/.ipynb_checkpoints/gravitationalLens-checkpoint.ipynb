## Finding strong lenses in astronomical images

An attempt was made to find strong lenses in images using a technique similar to PCA based lens finder. The main steps are to fit a model(Sersic or MOG's) to the galaxy profile and subract it - analyze the residual through a transformation to polar coordinates by looking for peaks in median row intensity values.

Primarily four methods were explored.

<b>a. Method 1</b> polar.py - implements the procedure of transforming the image to polar coordinates about the galaxy's center and plotting the median row intensity values. A sersic profile is fitted using scipy.optimize library.

Public Functions

plot_polar_image(): Plots an image reprojected into polar coordinates.

index_coords(): Creates x & y coords for the indicies in a numpy array

cart2polar(): Transforms from cartesian to polar space.

polar2cart(): Transforms from polar to cartesian space.

reproject_image_into_polar(): Projects a numpy array into a polar coordinate system.

sersic(): Template sersic function.

get_statistics(): Returns the row median and mean.

fit_profile(): Returns a sersic model fit to the data.

Summary:
1. Read image usinf scipy.misc.imread()
2. Transform to polar coordinates about the center.
3. Get the statistics.
4. Use fit_profile() function and plot results.

<b>b. Method 2 </b>In second method - convert an image into polar coordinates about a 'chosen pixel', then find the row mean in the transform for each row. On these row means, fit a serisic profile, and the 'real center' will be the one which has maximum correlation between the fitted seresic profile and the row means. Vary the 'chosen pixel' for all image pixels. This method returns incorrect centers. The center returned by this method is the green dot towards the right.

The aim of this technique was to combine the use of distance correlation and fitting 1D sersics to find the center of the foreground strong lensing galaxy.

<b>c. Method 3 </b> Direct 2D fits - For these many options were explored including PyMix, PyMc, Galfit, scipy optimize, griffin shapelets and astropy and others. The objective is to fit a 2D profile directly and subtract it to get the residual. Scipy optimize was tried for gaussian and sersic fuctions. While the gaussian functions converged for a good initial estimate - the sersic profile could not be converged.

```python
Image(filename='scipy_gauss_fit.jpg')
```

The right figure shows the original image - the left one is after subtracting a 2D gaussian. Since a gaussian decays very slowly - a single gaussian cannot accurately fit a decaying exponental (like sersic). Even though there is some reduction in the intensity it is not perfect.

Another option was to try a mixture of gaussians through PyMix. A drawback with PyMix is that it does not tell the optinum number of gaussians to estimate our data with. Pymc does not have MOG's. Galfit was tried but its source code is not available. Shapelets by griffin were tried. Astropy fittings did not have a sersic function implementation.

#### d. Mixture of gaussians

Due to the drawbacks of fitting a single 2D gaussian and the relative intractability of 2D sersic functions - another solution is fitting a mixture of gaussians. A superposition of gaussians can be used to represent varied profiles which can mimic exponential, sersic lux and luv profiles.

```python
    Image(filename='sample_fit_1.png')
```

We can use either Scipy or PyMix to deal with mixture of gaussians. The above example shows how expectation maximization technique in PyMix can be used to estimate a mixture of gaussians. 

The main steps are:

\# Create mixture of models

n1 = mixture.NormalDistribution(-1,1)

n2 = mixture.NormalDistribution(1,1)

m = mixture.MixtureModel(2,[0.5,0.5], [n1,n2])

\# perform expectation maximization

m.EM(data, 40, .1)

c1, c2 = m.components

p = m.pi

\# p[0] - relative fraction in component 1 

\# c1.distList[0].mu - a handle on the mu and sigma of the fit

\# c1.distList[0].sigma


According to "Replacing standard galaxy profiles with mixtures of Gaussians" - David W. Hogg & Dustin Lang, we can also model other complex functions by a mixture of gaussians - which makes them mathematically more tractable. The paper includes their results for exponential, lux and luv profiles. A 1-D approximation to an exponential by the superpostion of 4 gaussians is described.

```python
    Image(filename='sample_mix_1.png')
```

In the above image - the blue profile mimics an exponential created by the superposition of 4 gaussian profiles with specific weights and variances. The mean, variance and the number of gaussian functions are taken from David W. Hogg & Dustin Lang's paper. The same approach can be extended to 2D as well. Sample functions in mog.py mimic the lux, luv and exponential 2D gaussian profiles and are handled with scipy routines. A mixture of gaussians is fit to the image for removing the foreground galaxy.

<u>Functions</u>

twoD_Gaussian() - Generalized gaussian with rotation.

normpdf_2D() - Gaussian for 2D data (no theta).

sersic() - A sersic function for fitting (Does not converge)

lux_mix_2D_4() - A mixture of 4 gaussians mimicing lux profile

lux_mix_2D_6() - A mixture of 6 gaussians mimicing lux profile

lux_mix_2D_8() - A mixture of 8 gaussians mimicing lux profile

exp_mix_2D_4() - A mixture of 4 gaussians mimicing exp profile

exp_mix_2D_6() - A mixture of 6 gaussians mimicing exp profile

exp_mix_2D_8() - A mixture of 8 gaussians mimicing exp profile

luv_mix_2D_8() - A mixture of 8 gaussians mimicing luv profile

luv_mix_2D_10() - A mixture of 10 gaussians mimicing luv profile

The results are shown below - the luv_mix_2D_10 after optimization seems to remove the foreground galaxy most effectively. The first subgraph shows the original image, the second shows the model fitted and the third the image after model subtraction. The program might cause the fitting to be non-optimal in case of a bad starting estimate (stuck in local minima) or take a long time to compete otherwise. While using the routines it is advisable  to run them multiple times on different initial estimates - or use a good heuristic for determining the center. OptimalParametersNotFound exceptions should be handled in this case.
