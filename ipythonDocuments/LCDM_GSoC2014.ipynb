#### a. Coding the distance correlation method

The project uses distance correlation technique to estimate the time delay between two curves. "dcor.py" file contains the class implementing it. It takes as input two vectors of equal length and returns the corresponding distance correlation.

```python
import math
import numpy as np

'''
A Class for calculating the distance correlation beween two vectors of 'same' size.
Constructor    :     dcor(X,Y)                Returns an object of dcor class. X and Y are input vectors.
Functions      :     find_correlation()        Returns the distance correlation.
'''

class dcor:
    def __init__(self, X, Y):
        X_Mat = np.zeros((len(X),len(X)))               # X distance matrix - for all pairwise distances in X 
        Y_Mat = np.zeros((len(X),len(X)))               # Y distance matrix - for all pairwise distances in Y
        D_X_Mat = np.zeros((len(X),len(X)))             # X doubly centered distance matrix
        D_Y_Mat = np.zeros((len(X),len(X)))             # Y doubly centered distance matrix
        X_Row_Sum = np.zeros(len(X))                    # i'th Row sum of X distance matrix
        Y_Row_Sum = np.zeros(len(X))                    # i'th Row sum of Y distance matrix
        X_Col_Sum = np.zeros(len(X))                    # i'th Col sum of X distance matrix
        Y_Col_Sum = np.zeros(len(X))                    # i'th Col sum of Y distance matrix
        self.dCor = -1                                  # class variable - Correlation between X and Y

        if(len(X)!=len(Y)):                             # Check for same size of X and Y
            print "Distance Correlation: Input vectors must be of same size -"+str(len(X))+" != "+str(len(Y))
            return -1    
        
        if(len(X)==0):                                  # If Empty
            print "Distance Correlation: Input vector empty"
            return -1        
        
        for i in range (0,len(X)):
            sum_r_x = 0
            sum_r_y = 0            
            for j in range (0,len(X)):
                sum_r_x = sum_r_x + abs( X[i] - X[j])            # Adding sum to find the row mean 
                sum_r_y = sum_r_y + abs( Y[i] - Y[j])                 
                X_Mat[i][j] = ( abs( X[i] - X[j])  )            # Eucledian norm equivalent to modulus here
                Y_Mat[i][j] = ( abs( Y[i] - Y[j])  )
            
            X_Row_Sum[i] = sum_r_x     
            Y_Row_Sum[i] = sum_r_y     

        for i in range (0,len(X)):                                # Finding the column mean
            sum_c_x = 0
            sum_c_y = 0            
            for j in range (0, len(X)):
                 sum_c_x += X_Mat[j][i]        
                 sum_c_y += Y_Mat[j][i]
            X_Col_Sum[i] = sum_c_x
            Y_Col_Sum[i] = sum_c_y             


        X_mean = sum(X_Row_Sum) / (len(X) ** 2.0)                    # Overall mean equivalent to sum of all row sums  
        Y_mean = sum(Y_Row_Sum) / (len(X) ** 2.0)                    # divided by total number of elements i.e. n^2

        for i in range (0,len(X)):
            for j in range (0, len(X)):                             # Forming doubly centered distance matrix - cast len(X) to float for div.
                D_X_Mat[i][j] = ( X_Mat[i][j] - X_Row_Sum[i] / float(len(X)) - X_Col_Sum[j] / float(len(X)) + X_mean )
                D_Y_Mat[i][j] = ( Y_Mat[i][j] - Y_Row_Sum[i] / float(len(X)) - Y_Col_Sum[j] / float(len(X)) + Y_mean )

        Sum_Cov = 0
        Sum_Var_X = 0
        Sum_Var_Y = 0
        for i in range (0,len(X)):
            for j in range (0, len(X)):
                Sum_Cov     = Sum_Cov     + D_X_Mat[i][j] * D_Y_Mat[i][j]
                Sum_Var_X    = Sum_Var_X    + D_X_Mat[i][j] * D_X_Mat[i][j]
                Sum_Var_Y     = Sum_Var_Y    + D_Y_Mat[i][j] * D_Y_Mat[i][j]
        
        dCov       = Sum_Cov   / (len(X) ** 2.0)
        dVar_X     = Sum_Var_X / (len(X) ** 2.0)
        dVar_Y     = Sum_Var_Y / (len(X) ** 2.0)
     
        self.dCor = dCov / math.sqrt( dVar_X * dVar_Y)

    def find_correlation(self):
        return self.dCor             
```

The code uses numpy arrays for faster matrix operations. The following examples shows a simple use case.

```python
'''Example'''
A = [3,4,5,6]
B = [4,5,6,7]
print dcor(A,B).find_correlation()
```

Distance correlation is insensitive to uniform scale changes in a direction - hence the correlation between A and B is 1.

#### b. Finding shifts in lightcurves


The objective is to input a file containing data of two or more curves and find the shift between them. Correlation between curves is calculated for each shift iteration and the value corresponding to maximum correlation is taken as the shift between curves. The light curves are input as plain text files with each row in the format 

{mhjd-date [Mag(Curve i) Err(Curve i)] for i curves}

corr_plot.py has functions to load, segment, plot and estimate the shift between two light curves.

For loading the data - use the load_data() function. It loads the first argument passed from the terminal and returns the curve and the time data. For plotting and finding correlation, extract the 2 light curves of interest. The spline fitting can be done with fit_curve function. MAX_GAP parameter contols maximum days of non observance after which segments will be disjointed. plot_data_points function plots the input data points. plot_segment function plots a segment object. Both functions have a parameter to offset the plot by the given units in Y direction. A sample plotting routine will look like:	

time, Lcurves = load_data() 

LC_1, LC_2, T_Segments = fit_curve(time, lc_1, lc_2, MAX_GAP) #lc_1 and lc_2 are the curves points of interest extracted from Lcurves

plot_data_points(time, lc_1, lc_2, avg[x], avg[x+1])

To find the shift between two curves use - find_shift(LC_1, LC_2, Max_Shift, Shift_Step) which returns the shift in mhjd days.
Here LC_1 and LC_2 are two lists of 'SEGMENT' objects. The 'Max_Shift' specifies the maximum amount of days one curve can be shifted with respect to another. 'Shift_Step' controls the granuality of successive increments. 

Two types of fitting have been implemented - the univariate spline fitting to the segments and a cubic polynomial fitting to segments. The cubic polynomial fitting can be found in corr_plot_diff_algo.py

A description of the classes and functions is as follows:

<b>class Segment(object):</b> Class for light curve segment - used for simulating a contiguous lc section.

<u>Members:</u> 

intensity_function - Interpolating function of the segment

time_start - Starting time of the segment (minimum of time_data)

time_end - End time of the segment (maximum of time_data)	

time_data - List of points in time domain where the segment was sampled

intensity - Intensity values corresponding to time_data

time_data and intensity are stored to re-evaluate the fitting function after a transformation on data points.

<u>Functions:</u>

shift_segment(days) - Shifts a segment by the given number of mhjd days. The interpolating function is re-evaluated.

intersection(segment) - Returns a list of intensities corresponding to temporal overlap of the passed segment with the segment it is called upon.

getIntensity(time) - Returns the intensitiy at a given time by evaluating the interpolating function.

getStartTime() - Returns the start time of the segment.

getEndTime() - Returns the end time of a segment.

getSegmentLength() - Returns the length of the segment.

display() - Displayes the parameters of the light segment.

<b>Public Functions</b>

load_data() - Returns a lists of elements having the format [ Time, [[Lightcurve Ai, Error in Ai] for i curves]]. Header of input file is ignored (first two lines). 
	
plot_data_points(time, lc_a, lc_b, offset1, offset2): Plots a pair of LC data points offset in amplitude by the given amounts.

plot_segment(LC, avg): Plots a light curve segment offset by a given amount.

show_plots(): Displays all the plots.

find_shift(LC_1, LC_2, Max_Shift, Shift_Step): Finds the shift between two light curves. This function shifts all the segments in a light curve by an amount 'Shift_Step' till the Max_Shift and for each configuration finds the intersections between the curve segments. The correlation between them is recorded and the shift corresponding to maximum correlation is output.

fit_curve(time, lc_a, lc_b, MAX_GAP): Takes a pair of raw light curve intensity values and returns a list of 'Segment' Objects - each Segment object has its own interpolating function. MAX_GAP corresponds to the number of non observational days after which the segment will terminate. Each segment should have a minimum of 5 observational points.

<u>Summary:</u>

1. Load data using 'load_data()' function.
2. Extract the light curves of interest.
3. Fit segments using 'fit_curve()' function on the extracted light curves.
4. Find the shift using 'find_shift()' function on the light segments.
5. Plot using 'plot_segment()', 'plot_data_points()' and 'show_plots()' function.

## Testing on Artificial Light Curves

Testing on artificial light curves was done and the code returned good estimates of the shift - excluding the cases where the seasonal effects wiped more than about 50% of the observations on large shifts relative to the light curve length (which is tricky anyways). Care should always be taken that high shifts in two light curves will include less and less overlap points - causing the correlation to be higher sometimes; giving false results. Hence the maximum allowed shift should be kept small so that sufficient number of overlap points are there as a meaure of confidence. 

Some test cases are:

a. Small shift with less seasonal effects - a file with a shift of 9 days is generated which is predicted correctly.
